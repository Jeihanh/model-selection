MLE Estimators (Data 1) 

```python
from scipy.stats import halfnorm, lognorm, levy, expon, gumbel_r, cauchy, gamma, pareto

def fit_distributions(data):
    # Half-normal distribution 
    halfnorm_param = np.std(data['Value'])

    # Lognormal distribution
    shape, loc, scale = lognorm.fit(data['Value'])
    sigma_log = shape
    mu_log = np.log(scale)  

    # Levy distribution
    levy_params = levy.fit(data['Value'])

    # Exponential distribution
    lambda_exp = 1 / np.mean(data['Value'])

    # Pareto distribution
    pareto_params = pareto.fit(data['Value'], floc=0)

    # Gamma distribution
    gamma_params = gamma.fit(data['Value'], floc=0)

    # Gumbel distribution
    gumbel_params = gumbel_r.fit(data['Value'])

    # Cauchy distribution
    cauchy_params = cauchy.fit(data['Value'])

    # Print results
    print("Half-normal parameter:", halfnorm_param)
    print("Sigma, Mu (Lognormal):", sigma_log, mu_log)
    print("Levy Parameters:", levy_params)
    print("Lambda (Exponential):", lambda_exp)
    print("Pareto Parameters:", pareto_params)
    print("Gamma Parameters:", gamma_params)
    print("Gumbel Parameters:", gumbel_params)
    print("Cauchy Parameters:", cauchy_params)
```

```python
# Apply the function to each dataset
print("Fitting distributions for Data1:")
fit_distributions(data1)
print("\nFitting distributions for Data2:")
fit_distributions(data2)
```
